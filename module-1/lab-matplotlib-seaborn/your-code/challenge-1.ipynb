#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)

fig = plt.figure(figsize=(12,6))

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("X vs. Y")
plt.subplot(1, 2, 2)
plt.plot(x, z)
plt.xlabel("X")
plt.ylabel("Z")
plt.title("X vs. Z")
plt.show()


```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
fig, a = plt.subplots(1, 2)
fig.suptitle('The two plots')
a[0].plot(x, y)
a[1].plot(x,z)
plt.show()
```

#### In the vehicles data set that you have downloaded, use the vehicles.csv file. In this exercise we will conduct some exploratory data analysis using one plot each of scatter plot, box plot, histogram, and bar chart. 

Scatter Plot

Please provide a scatter plot between "Combined MPG" as X variable and 
"Highway MPG" as Y variable

```python
# your code here
df = pd.read_csv("vehicles.csv", index_col=0)
df.head()
```

```python
plt.figure(figsize=(9,9))
plt.scatter(df["Combined MPG"], df["Highway MPG"])

plt.xlabel("Combined MPG")
plt.ylabel("Highway MPG")
plt.title("Scatter plot")
plt.show()
```

Box Whisker Plot

Please provide a box plot of the variable "CO2 Emission Grams/mile"

```python
# your code here
plt.figure(figsize=(4,8))
plt.boxplot(df["CO2 Emission Grams/Mile"])
plt.title("Boxplot")
plt.xlabel("All vehicles in the database")
plt.ylabel("CO2 Emission (Grams/Mile)")
plt.show()
```

Histogram

Please provide a histogram of the Fuel Barrels/Year

```python
# your code here
plt.figure(figsize=(8,4))
plt.hist(df["Fuel Barrels/Year"])
plt.title("Histogram")
plt.xlabel("All vehicles in the database")
plt.ylabel("Fuel Barrels/Year")
plt.grid(True)
plt.show()
```

Bar Chart

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
# your code here ,Fuel Type,City MPG

dic= {k:df[df["Fuel Type"] == k]["City MPG"].mean() for k in df["Fuel Type"].unique()}

plt.figure(figsize=(14,6))
plt.grid(True)
plt.bar(range(len(dic)), list(dic.values()), align='center')
plt.xticks(range(len(dic)), list(dic.keys()), rotation=45,ha='right')
plt.ylabel("Mean City MPG")
plt.title("Bar plot")
plt.show()
```
