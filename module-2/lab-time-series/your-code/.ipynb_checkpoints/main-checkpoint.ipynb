The next step will be to test both variables for stationarity. Perform the Augmented Dickey Fuller test on both variables below.

```python
# Your code here:

# h0 = La hipótesis nula establece que existe la presencia de una raíz unitaria.
# h1 = La hipótesis alternativa establece que no hay raíz unitaria. En otras palabras, la estacionariedad existe.

fuller = adfuller(df["userAcceleration.x"])

fuller
```

```python
alpha = 0.05

valor_critico_datos = fuller[0]
valor_p = fuller[1]
rezagos_utilizados = fuller[2]
n_observaciones = fuller[3]
valores_t = fuller[4]

if valor_p < alpha:
    print("Se rechaza la hipótesis nula. Se acepta la hipostesis alternativa. ¡¡HABEMUS ESTACIONARIEDAD!!")
elif valor_p > alpha:
    print("Se acepta la hipótesis nula")
```

What are your conclusions from this test?

```python
# Your conclusions here:

# Las propiedades de la serie no varían con respecto al tiempo. En otras palabras, 
# su variación (la forma en la que cambia) no cambia en función del tiempo.
```
