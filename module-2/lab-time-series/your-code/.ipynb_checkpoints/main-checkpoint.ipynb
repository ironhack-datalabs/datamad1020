# Challenge 2 - Modelling the Data

To model our data, we should look at a few assumptions. First, let's plot the `lag_plot` to detect any autocorrelation. Do this for `userAcceleration.x`

```python
rcParams['figure.figsize'] = 5, 5
```

```python
pd.plotting.lag_plot(sensor['userAcceleration.x'], c = 'purple')
```

Create a lag plot for `rotationRate.x`

```python
pd.plotting.lag_plot(sensor['rotationRate.x'], c = 'yellow')
```

What are your conclusions from both visualizations?

#### - Due to the pattern displayed, both plots are positive (values increase together), and with a strong correlation.
#### - In the plot regarding userAcceleration.x, the values are much more concentrated and go from min = -1.5 to max = 1.5 in both axis. 
#### - On the other hand, in the  plot for rotationRate.x, the values are more spread in both axis going from min = -4 to max = 3.

The next step will be to test both variables for stationarity. Perform the Augmented Dickey Fuller test on both variables below.

```python
#Documentation; statsmodels.tsa.stattools.adfuller(x, maxlag=None, regression='c', autolag='AIC', store=False, regresults=False)

# ADF Test
ADF_1 =  sm.tsa.stattools.adfuller(sensor['userAcceleration.x'],autolag = 'AIC')
print(f'ADF Statistic: {ADF_1[0]}')
print(f'n_lags: {ADF_1[1]}')
print(f'p-value: {ADF_1[1]}')
for key, value in ADF_1[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}')  
```

```python
ADF_2 =  sm.tsa.stattools.adfuller(sensor['rotationRate.x'],autolag = 'AIC')
print(f'ADF Statistic: {ADF_2[0]}')
print(f'n_lags: {ADF_2[1]}')
print(f'p-value: {ADF_2[1]}')
for key, value in ADF_2[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}')  
```

What are your conclusions from this test?

#### In theory: When the test statistic is lower than the critical value shown, you reject the null hypothesis and infer that the time series is stationary. Therefore, ADF_1 and ADF_2 are stationary
