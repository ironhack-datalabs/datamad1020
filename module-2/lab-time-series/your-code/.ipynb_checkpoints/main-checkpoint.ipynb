# Challenge 2 - Modelling the Data

To model our data, we should look at a few assumptions. First, let's plot the `lag_plot` to detect any autocorrelation. Do this for `userAcceleration.x`

```python
plt.figure(figsize=(10,4))
grafica = pd.plotting.lag_plot(sensor["userAcceleration.x"])
```

Create a lag plot for `rotationRate.x`

```python
plt.figure(figsize=(10,4))
grafica2 = pd.plotting.lag_plot(sensor["rotationRate.x"])

```

What are your conclusions from both visualizations?

**En ambas graficas tenemos una correlacin positiva relativamente fuerte, ya que tiene
una tendencia creciente y con los puntos tan apretados sugiere una correlacion fuerte, como ya explique antes.
Si cabe desatacar que la correlacion de los graficos se vuelve un poco ms debil al final del tiempo**


The next step will be to test both variables for stationarity. Perform the Augmented Dickey Fuller test on both variables below.

```python
from statsmodels.tsa.stattools import adfuller
aceleracion_fuller = adfuller(sensor["userAcceleration.x"])
p_value = aceleracion_fuller[1]
print(p_value)
print("-------------------------")

rotation = adfuller(sensor["rotationRate.x"])
p_value_2 = rotation[1]
print(p_value_2) 
```

What are your conclusions from this test?

**con un alpha elegido de 0.05 y al ser el pvalue menor que sta, podemos rechazar
la hipotesis nula de que la variable no es estacionaria, es decir, casi con toda probabilidad podemos afirmar que 
la variable es estacionaria, es decir, la media, la varianza, y la autocorrelacin permanecen constantes a lo largo del tiempo.**

**igual que la anterior, un pvalue muy bajo que rechaza la hipotesis nula de que no es estacionaria**

Finally, we'll create an ARMA model for `userAcceleration.x`. Load the `ARMA` function from `statsmodels`. The order of the model is (2, 1). Split the data to train and test. Use the last 10 observations as the test set and all other observations as the training set. 

```python
from  statsmodels.tsa.arima_model import ARIMA

test = sensor.iloc[-10:]

train = sensor.iloc[:-10]
plt.figure(figsize=(12,4))
train['userAcceleration.x'].plot()
test['userAcceleration.x'].plot()
```

```python
model=ARIMA(train['userAcceleration.x'],order=(2,0,1))
model=model.fit()

```

```python
predictions = model.predict(len(sensor["userAcceleration.x"])-10, len(sensor["userAcceleration.x"])-1)
pd.DataFrame({'observed':sensor["userAcceleration.x"][-10:], 'predicted':predictions})
```

To compare our predictions with the observed data, we can compute the RMSE (Root Mean Squared Error) from the submodule `statsmodels.tools.eval_measures`. You can read more about this function [here](https://www.statsmodels.org/dev/generated/statsmodels.tools.eval_measures.rmse.html). Compute the RMSE for the last 10 rows of the data by comparing the observed and predicted data for the `userAcceleration.x` column.

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
test['userAcceleration.x'].mean()
rmse=sqrt(mean_squared_error(predictions,test['userAcceleration.x']))
print(rmse)
#tenemos un error del 22.5%, no esta mal.. pero podria ser mejor.
```
