Our next step is to decompose the time series and evaluate the patterns in the data. Load the `statsmodels.api` submodule and plot the decomposed plot of `userAcceleration.x`. Set `freq=60` in the `seasonal_decompose` function. Your graph should look like the one below.

The next step will be to test both variables for stationarity. Perform the Augmented Dickey Fuller test on both variables below.

```python
# Your code here:
from statsmodels.tsa.stattools import adfuller

#adfuller(occupancy.CO2)[1]
print("User Acceleration X: ", adfuller(sensor['userAcceleration.x'])[1])
print("Rotation Rate X: ", adfuller(sensor['rotationRate.x'])[1])
```

What are your conclusions from this test?

```python
# Your conclusions here:

# P_value are in both cases less than 0.05. Therefore, with a 95% confidence interval,
# we can reject the null hypotesis(H0). The data doesn't have unit root, both series 
# are stationary.
```

Finally, we'll create an ARMA model for `userAcceleration.x`. Load the `ARMA` function from `statsmodels`. The order of the model is (2, 1). Split the data to train and test. Use the last 10 observations as the test set and all other observations as the training set. 

```python
# Your code here:
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(sensor['userAcceleration.x'], order=(2,0, 1))
model_fit = model.fit()
predictions = model_fit.predict(len(sensor['userAcceleration.x'])-10, len(sensor['userAcceleration.x'])-1)
print(predictions)

# pd.DataFrame({'observed':sensor['userAcceleration.x'][-10:], 'predicted':predictions[:-10]})

#No he sabido como continuar, lo encuentro muy lioso y llega un momento que no sé lo que tengo que hacer ni como continuar....
```
